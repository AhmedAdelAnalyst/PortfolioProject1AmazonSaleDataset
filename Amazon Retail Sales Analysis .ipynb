```python
#drop unnamed colum with 49050 null values 
df=df.drop(['Unnamed: 22'], axis=1)
df
```

```python
#fill null values in each colume
df['Courier Status'] = df['Courier Status'].fillna(0)
df['currency'] = df['currency'].fillna('INR')
df['Amount'] = df['Amount'].fillna(0)
df['ship-city'] = df['ship-city'].fillna('Unknown')
df['ship-state'] = df['ship-state'].fillna('Unknown')
df['ship-postal-code'] = df['ship-postal-code'].fillna('Unknown')
df['ship-country'] = df['ship-country'].fillna('Unknown')
df['promotion-ids'] = df['promotion-ids'].fillna(0)
df['fulfilled-by'] = df['fulfilled-by'].fillna(0)
df.isna().sum()
```

```python
# Convert date column to datetime
df['Date']= pd.to_datetime(df['Date'])
```

```python
#drop 2022-03-31 from the dataset because its the only available day in march in the dataset
df.drop(df.index[df['Date'] == '2022-03-31'], inplace = True)
```

```python
pd.set_option('display.max_columns', None)
```

```python
# Create a new `month` column
df['month'] = df['Date'].dt.month
df.head()
```

```python
# Calculate total number of sales per month
df.groupby(['month']).sum().sort_values('Qty', ascending=False).head(12)
```

```python
# Create a new `month_txt` column.
df['month_txt'] = df['Date'].dt.month_name().str.slice(stop=3)
df.head()
```

Question 1 : What was the best month for sales? How much was earned that month?

```python
# Create a new helper dataframe for plotting.
df_by_month = df.groupby(['month','month_txt']).sum().sort_values('month', ascending=True).head(12).reset_index()
df_by_month
```

```python
plt.bar(x=df_by_month['month_txt'],height= df_by_month['Qty'], label="Qty")
plt.plot()

plt.xlabel("Months(2022)")
plt.ylabel("Qty")
plt.title("Qty by months")
plt.legend()
plt.show()
```

Question 2 : what week with the highest revenue ?

```python
# Create four new columns.
df['week'] = df['Date'].dt.strftime('%Y-W%V')
df['month'] = df['Date'].dt.strftime('%Y-%m')
df['quarter'] = df['Date'].dt.to_period('Q').dt.strftime('%Y-Q%q')
df['year'] = df['Date'].dt.strftime('%Y')
```

```python
df.head(10)
```

```python
#divide qty by 1000000
df_by_week = df['Amount'].div(1000000)
df_by_week .head()
```

```python
# Group data by week and sum.
df_by_week = df.groupby(['week']).sum().reset_index()

# Format as text, in millions.
df_by_week['Amount_formatted'] = df_by_week['Amount'].div(1000000).round(1).astype(str) + 'M'

df_by_week.head(20)
```

```python
def addlabels(x, y, labels):
    '''
    Iterates over data and plots text labels above each bar of bar graph.
    '''
    for i in range(len(x)):
        plt.text(i, y[i], labels[i], ha = 'center', va = 'bottom')
```

```python
plt.figure(figsize = (15, 5))
plt.bar(x = df_by_week['week'], height = df_by_week['Amount'])
addlabels(df_by_week['week'], df_by_week['Amount'], df_by_week['Amount_formatted'])
plt.plot()
plt.xlabel('WEEK')
plt.ylabel('AMOUNT')
plt.title('AMOUNT PER WEEK (2022)')
plt.show()
```

 Question 3 : what state with the highest number of shipments ?

```python
#group the data by state and sort by qty
df_by_state = df.groupby(['ship-state']).sum().sort_values('Qty', ascending=False).head(12).reset_index()
df_by_state
```

```python
keys = [state for state, df in df.groupby(['ship-state'])]
plt.figure(figsize = (16, 6))
plt.bar(keys,df.groupby(['ship-state']).sum()['Qty'])
plt.ylabel('QTY')
plt.xlabel('STATE')
plt.xticks(keys, rotation='vertical', size=8)
plt.show()
```

 Question 4 : what category sales the most ?
