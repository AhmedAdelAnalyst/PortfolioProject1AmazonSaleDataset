1. Which Category is Best Selling?

```python
# Group sales by category
category_analysis = pd.DataFrame(df.groupby(['Category'])[['Sales']].sum())
category_analysis
```

```python
# Plot Bar Plot
sns.set_theme(style="whitegrid")
sns.barplot(data = category_analysis, x = category_analysis.index, y = category_analysis.Sales, palette = "muted")
plt.title("sales by category")
plt.show()
```

2. What are the Best Selling Sub-Category?

```python
# Group by sub-category
subcat_analysis = pd.DataFrame(df.groupby(['Sub-Category'])[['Sales']].sum())# Sort by descending order according to sales
subcat_sales = pd.DataFrame(subcat_analysis.sort_values('Sales', ascending = False))
subcat_sales
```

```python
# Plot Bar Plot for Top Selling Sub-Category
sns.set_theme(style="whitegrid")
sns.barplot(data = subcat_sales, y = subcat_sales.index, x = subcat_sales.Sales, palette = "muted")
plt.title("Top Selling Sub-Category")
plt.show()
```

```python
segment_analysis = pd.DataFrame(df.groupby(['Segment'])[['Sales']].sum())
segment_analysis
```

```python
# Plot Bar Plot
sns.set_theme(style="whitegrid")
sns.barplot(data = segment_analysis, x = segment_analysis.index, y = segment_analysis.Sales, palette = "rocket")
plt.title("Customer Segment Sales")
plt.show()
```

Which is the Preferred Ship Mode?

```python
shipmode_analysis = pd.DataFrame(df.groupby(['Ship Mode'])[['Sales']].sum())
shipmode_analysis
```

```python
# Plot Bar Plot
sns.set_theme(style="whitegrid")
sns.barplot(data = shipmode_analysis, x = shipmode_analysis.index, y = shipmode_analysis.Sales, palette = "rocket")
plt.title("Customer Segment Sales")
plt.show()
```

6. Which Region is the Highest In Sales?

```python
region_analysis = pd.DataFrame(df.groupby(['Region'])['Sales'].sum().reset_index())
region_analysis
```

```python
# Plot Pie Chart
explode = [0, 0, 0, 0.1]
plt.pie(region_analysis.Sales, labels = region_analysis.Region , startangle = 90, autopct = "%1.0f%%", explode = explode, shadow = True)
plt.title("Most Profitable by Region")
plt.show()
```
